```python
from bs4 import BeautifulSoup
import requests
import random
import time
import pandas as pd
from ua_info import ua_list
```

```python
def get_html(url, param):
    # 重构请求头，使用自定义UA标识池
    headers = {'User-Agent': random.choice(ua_list)}
    response = requests.get(url=url, params=param, headers=headers)
    # 每爬取一个页面随机休眠1-2秒钟的时间
    time.sleep(random.randint(1, 2))
    return response
```

```python
def save_csv(dic, filename):
    df = pd.DataFrame(dic)
    df.to_excel(filename, index=False)
```

```python
def parse_html(response):
    # 生成BeautifulSoup对象
    soup = BeautifulSoup(response.text, 'html.parser')
    post_list = soup.find_all("li", {'class': 'j_thread_list clearfix thread_item_box'})
    lt = []
    for posts in post_list:
        post = {}
        try:
            post['标题'] = posts.find('a',{'class':'j_th_tit'}).get_text()
            post['作者'] = posts.find('span',{'class':'frs-author-name-wrap'}).get_text()
            post['内容'] = posts.find('div', {'class': "threadlist_abs threadlist_abs_onlyline"}).get_text()
            print(post)
        except Exception as e:
            print(e)
            continue
        # 数据清洗
        if "回复" in post['标题']:
            pass
        elif "转贴" in post['标题']:
            pass
        else:
            # 列表拼接成字典
            lt.append(post)
    return lt
```

```python
url = 'https://tieba.baidu.com/f?'  # 目标url
kw = '孙笑川'
# content = input("请输入要查询的内容")  # 内容
page_start = 1  # 页码
page_end = 2  # 页码
filename = '孙笑川'  # 文件名
lt2 = []  # 新建空列表用于存储数据

param = {
    'ie': 'utf-8',  # 编码方式
    'kw': kw,  # 吧名
    # 'qw': content,  # 内容
    'pn': 1  # 页码
}
res = get_html(url, param)
parse_html(res)
```

```python
def run():
    url = 'https://tieba.baidu.com/f?'  # 目标url
    # content = input("请输入要查询的内容")  # 内容
    lt2 = []  # 新建空列表用于存储数据
    for page in range(0, 1, 1):
        param = {
            'ie': 'utf-8',  # 编码方式
            'kw': '孙笑川',  # 吧名
            # 'qw': content,  # 内容
            'pn': page*50  # 页码
        }
        res = get_html(url, param)
        lt1 = parse_html(res)
        # 列表拼接
        lt2.extend(lt1)
    save_csv(lt2, filename)
```

```python
try:
    run()
except Exception as e:
    print("错误:", e)
```
